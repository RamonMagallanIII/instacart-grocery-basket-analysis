## Missing Data 

```python
# Checking for NaNs
df_cust.isnull().sum()
```

```python
# The nulls will be replaced with John if NaN and male, and Jane if NaN and female.
# Though this is about 5.5% of the customer dataframe, it's potential for linked row count in the merged dataframe is much greater.
df_cust.loc[df_cust['first_name'].isna() & (df_cust['gender'] == 'Male'), 'first_name'] = 'John'
```

```python
df_cust.loc[df_cust['first_name'].isna() & (df_cust['gender'] == 'Female'), 'first_name'] = 'Jane'
```

```python
# Checking results in column
df_cust['first_name'].isna().sum()
```

```python
# Checking results within the dataframe
df_cust.isnull().sum()
```

## Mixed-Type Data

```python
# Checking for mixed data
for col in df_cust.columns.tolist():
  weird = (df_cust[[col]].applymap(type) != df_cust[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_cust[weird]) > 0:
    print (col)
```

```python
# Check all columns for mixed data types using map()
for col in df_cust.columns:
    unique_types = df_cust[col].map(type).unique()
    if len(unique_types) > 1:
        print(f"âš ï¸ Mixed types in column '{col}': {unique_types}")
    else:
        print(f"âœ… '{col}' has consistent type: {unique_types[0]}")
```
