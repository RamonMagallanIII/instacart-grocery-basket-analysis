### 4.

```python
# Creating a column for the maximum order of each user_id
df_ords_prods['max_order'] = df_ords_prods.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# Confirming the max_order column was properly added
df_ords_prods.head(15)
```

```python
# Creating a flag for "loyal customer" based on their "max_order" being over 40
df_ords_prods.loc[df_ords_prods['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
# Creating a flag for "Regular customer" based on their "max_order" being under or equal to 40 and over 10
df_ords_prods.loc[(df_ords_prods['max_order'] <= 40 ) & (df_ords_prods['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
# Creating a flag for "New customer" based on their "max_order" being under or equal to 10
df_ords_prods.loc[df_ords_prods['max_order'] <=10., 'loyalty_flag'] = 'New customer'
```

```python
# Checking the frequency of our new "loyalty_flag" column
df_ords_prods['loyalty_flag'].value_counts(dropna = False)
```

```python
# Checking the results of our newly minted "loyalty_flag" column
df_ords_prods[['user_id','loyalty_flag','order_number']].head(60)
```

### 5.

```python
# Statistical analysis based on the "loyalty_flag"
df_ords_prods.groupby('loyalty_flag').agg({'prices': ['mean','min','max']})
```

New customers have the highest average spend per item, with regular customers trailing by only 6.01%. In contrast, loyal customers tend to purchase lower-priced items, spending approximately 21.88% less per item on average compared to new customers.

### 6.

```python
# Aggregating the average item price by user_id
df_ords_prods.groupby('user_id').agg({'prices': ['mean']})
```

```python
# Creating a column for the average item price by user_id
df_ords_prods['average_user_item_price'] = df_ords_prods.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Checking results of our new column
df_ords_prods.head(15)
```

```python
# Creating a flag for each user_id spending habit
df_ords_prods.loc[df_ords_prods['average_user_item_price'] >= 10, 'spending_habit'] = 'High spender'
```

```python
df_ords_prods.loc[df_ords_prods['average_user_item_price'] < 10, 'spending_habit'] = 'Low spender'
```

```python
# Checking the frequency of our new spending_habit flag
df_ords_prods['spending_habit'].value_counts(dropna = False)
```

```python
# Checking the results of our new spending_habit flag
df_ords_prods.sort_values('user_id')
```

```python
df_ords_prods.head(100)
```

### 7.

```python
# Aggregating the average days_since_prior_order by user_id
df_ords_prods.groupby('user_id').agg({'days_since_prior_order': ['mean']})
```

```python
# Creating a column for the average days_since_prior_order by user_id
df_ords_prods['average_days_since_prior_order'] = df_ords_prods.groupby(['user_id'])['days_since_prior_order'].transform(np.mean)
```

```python
# Checking results of our new column
df_ords_prods.head(15)
```

```python
# Creating a flag for each user_id by average_days_since_prior_order
df_ords_prods.loc[df_ords_prods['average_days_since_prior_order'] > 20, 'order_regularity'] = 'Non-frequent customer'
```

```python
df_ords_prods.loc[(df_ords_prods['average_days_since_prior_order'] > 10 ) & (df_ords_prods['average_days_since_prior_order'] <= 20), 'order_regularity'] = 'Regular customer'
```

```python
df_ords_prods.loc[df_ords_prods['average_days_since_prior_order'] <=10., 'order_regularity'] = 'Frequent customer'
```

```python
# Checking the frequency of our new order_regularity flag
df_ords_prods['order_regularity'].value_counts(dropna = False)
```

```python
# Checking the results of our new spending_habit flag
df_ords_prods.sort_values('user_id')
```
